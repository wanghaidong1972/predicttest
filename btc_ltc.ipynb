```python
import numpy as np
import pandas as pd
from scipy import stats
 
# グラフ描画
from matplotlib import pylab as plt
import seaborn as sns
%matplotlib inline
```

```python
btc = pd.read_csv("BITSTAMP_SPOT_BTC_USD.csv",index_col='time_period_start')

btc.reindex
btc.index.name='time'
btc.index = pd.to_datetime(btc.index)
```

```python
btc=btc[['price_close','volume_traded']]
```

```python
ltc = pd.read_csv("BITSTAMP_SPOT_LTC_USD.csv",index_col='time_period_start')
ltc.index.name='time'
ltc.index = pd.to_datetime(ltc.index)
ltc=ltc[['price_close','volume_traded']]
```

```python
df = pd.merge(btc, ltc, how = 'inner', left_index = True, right_index=True)
df.head()
```

```python
df['btc_diff'] = df.price_close_x - df.shift(1).price_close_x
df['ltc_diff'] = df.price_close_y - df.shift(1).price_close_y
df.head()
```

```python
df.dropna(how='any',inplace=True)
```

```python
df['ltc_lag'] = df.price_close_y - df.shift(-1).price_close_y
df.dropna(how='any',inplace=True)
```

```python
df['btc_lag'] = df.price_close_x - df.shift(-1).price_close_x
df.dropna(how='any',inplace=True)
df.head()
```
