```python
import numpy as np
import pandas as pd
from scipy import stats
 
# グラフ描画
from matplotlib import pylab as plt
import seaborn as sns
%matplotlib inline
```

```python
btc = pd.read_csv("BITSTAMP_SPOT_BTC_USD.csv",index_col='time_period_start')
```

```python
btc.head()
```

```python
btc.reindex
btc.index.name='time'
btc.index = pd.to_datetime(btc.index)
```

```python
btc.head()
```

```python
btc=btc[['price_close','volume_traded']]
btc.head()
```

```python
bch = pd.read_csv("BITSTAMP_SPOT_BCH_USD.csv",index_col='time_period_start')
bch.index.name='time'
bch.index = pd.to_datetime(bch.index)
bch=bch[['price_close','volume_traded']]
bch.head()
```

```python
df = pd.merge(btc, bch, how = 'inner', left_index = True, right_index=True)
df.head()
```

```python
df.corr()
```

```python
df['btc_diff'] = df.price_close_x - df.shift(1).price_close_x
df['bch_diff'] = df.price_close_y - df.shift(1).price_close_y
df.head()
```

```python
df.dropna(how='any',inplace=True)
df.head()
```

```python
df.corr()
```

```python
df['bch_lag'] = df.price_close_y - df.shift(-1).price_close_y
df.dropna(how='any',inplace=True)
df.head()
```

```python
df.corr()
```

```python
df['btc_lag'] = df.price_close_x - df.shift(-1).price_close_x
df.dropna(how='any',inplace=True)
df.head()
```
